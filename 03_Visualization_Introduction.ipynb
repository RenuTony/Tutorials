# Visualization with Jupyter

Following, there is small visualization example in a Jupyter Notebook. Two libraries are shown: matplotlib and Seaborn. For other accessible libraries see the [documentation](https://github.com/TSI-DIH).

## Importing Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading an Example Data Set

```python
tips = sns.load_dataset("tips")
```

## Visualize the Data

Some scatter plots implemented with Seaborn and matplotlib are shown here.

```python
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
plt.scatter('tip', 'total_bill', data = tips, marker = "x")
plt.ylabel('tip')
plt.xlabel('total_bill')
plt.show()
```

### Some more plots 

```python
sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",
           data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="bar", data=tips)
```
