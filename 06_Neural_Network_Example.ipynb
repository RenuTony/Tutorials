# Neural Network Example with Keras

Here, you can find an example of a multi-class classification with a small neural network constructed with the Keras library. As a data set the Iris data set from sklearn is used.

```python
import numpy as np
import keras
from sklearn import datasets
from sklearn import metrics
from numpy import argmax
from keras.utils import to_categorical
```

Loading the standard Iris dataset and having a look at the values

```python
iris = datasets.load_iris()
print(iris.data)
print(iris.target)
```

For classification we need the features (X) and the Flag (y) separated. Also, the outcome has to be categorical (one hot encoded). The data set is then split up into train and test to confirm that the neural network is functioning properly.

```python
X, y = iris.data, iris.target
```

```python
encoded = to_categorical(y)
print(encoded)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded, random_state=0)
```

Next is the training of the neural network with two hidden layers.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import numpy

# fix random seed for reproducibility
numpy.random.seed(3)

# create model
model = Sequential()
model.add(Dense(20, input_dim=4, kernel_initializer='glorot_uniform', activation='relu'))
Dropout(0.5)
model.add(Dense(10, kernel_initializer='glorot_uniform', activation='softmax'))
Dropout(0.5)
model.add(Dense(3, kernel_initializer='glorot_uniform', activation='softmax'))
# Compile model
optimizer = keras.optimizers.Adam(lr=0.001)
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
# Fit the model
model.fit(X_train, y_train, epochs=100, batch_size=20)
```

As you will see, the neural network was able to predict 97 % of the test data correctly.

```python
predictions = model.predict(X_test)
predictions = np.argmax(predictions, axis=1) 
y_test = np.argmax(y_test, axis=1)
print(metrics.accuracy_score(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
```
